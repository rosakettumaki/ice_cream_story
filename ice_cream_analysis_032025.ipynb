```python
from plotnine import *
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['svg.fonttype'] = 'none'
```

```python
df = pd.read_csv('ice_cream_consumption_top_10_countries.csv')
df.head()
```

```python
from plotnine import ggplot, aes, geom_bar

ggplot(df) + \
    aes(x='Country', y='Ice Cream Consumption (L)') + \
    geom_bar(stat="identity")
```

```python
from plotnine import ggplot, aes, geom_bar, theme_bw
import pandas as pd

# Ensure correct indentation
df['Country'] = pd.Categorical(df['Country'], 
                               categories=df.sort_values('Ice Cream Consumption (L)', ascending=False)['Country'],
                               ordered=True)

# Create the plot
chart = (
    ggplot(df) +
    aes(x='Country', y='Ice Cream Consumption (L)') +
    geom_bar(stat="identity") +
    theme_bw()
)

# Save the plot
chart.save('top10_icecream_consumers.svg')
```
